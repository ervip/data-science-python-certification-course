```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
"""
Load the file “zoo.data” and look at the info and the first five rows. The first 
column denotes the animal name and the last one specifies a high-level class 
for the corresponding animal.
"""

CSV_PATH = r'D:\CourseWork\data-science-python-certification-course\Assignments\10 Unsupervised Learning\Case Study III\resources\zoo.csv'
df = pd.read_csv(CSV_PATH)
df.head(5)
```

```python
# Find out the unique number of high-level classes
print(dict(zip(*np.unique(df['class_type'], return_counts=True))))
```

```python
"""
Use the 16-intermediate feature and perform an agglomerative clustering.
[ Hint: Refer to the agglomerative clustering module in scikit learn and set the 
number of clusters appropriately]
"""

agc = AgglomerativeClustering(n_clusters=4)
predicted = agc.fit_predict(df.iloc[:, 1:17])

print("Accuracy:", accuracy_score(df['class_type'], predicted))
print("MSE:", mean_squared_error(df['class_type'], predicted, squared=False))
```
