```python
"""
Domain
    Retail/Fashion

Focus
    Optimize sales

Business challenge/requirement
    Fyntra is the largest online clothing company in the USA. It sells clothing online, but 
    they also have in-store style and clothing advice sessions. Customers come into the 
    store, have sessions/meetings with a personal stylist, then can go home and order 
    either on a mobile app or website the clothes they want.
    The company wants to decide whether to focus the effort on the mobile app 
    experience or its website. As a drastic measure, it is also evaluating to shut down the 
    website. 
    You as an ML expert in the team will help the company make the right decision 
    
Key issues
    Establish a correlation among the parameters supplied in the data

Considerations
    NONE

Data volume
    Approx 500 records file FyntraCustomerData.csv (All data is fake so do not worry 
    about privacy)

Additional information
    - NA

Business benefits
    An increase in profits as the focus on the optimal sales channel will result in a higher 
    top line and the higher bottom line
"""
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
CSV_PATH = r'D:\CourseWork\data-science-python-certification-course\Assignments\06 Introduction to Machine Learning with Python\Case Study III\resources\FyntraCustomerData.csv'
df = pd.read_csv(CSV_PATH)
```

```python
"""
Compute -- Use seaborn to create a jointplot to compare the Time on Website 
and Yearly Amount Spent columns. Is there a correlation?
"""

tdf = df[["Time_on_Website", "Yearly_Amount_Spent"]]

sns.set(color_codes=True)
sns.jointplot(tdf, x="Time_on_Website", y="Yearly_Amount_Spent")
```

```python
# Is there a correlation? : Both have strong correlation

sns.heatmap(tdf.corr())
```

```python
"""
Compute - Do the same as above but now with Time on App and Yearly 
Amount Spent. Is this correlation stronger than 1st One?
"""

tdf = df[["Time_on_App", "Yearly_Amount_Spent"]]

sns.set(color_codes=True)
sns.jointplot(tdf, x="Time_on_App", y="Yearly_Amount_Spent")
```

```python
# Is this correlation stronger than 1st One?

sns.heatmap(tdf.corr())
```

```python
"""
Compute -- Explore types of relationships across the entire data set using 
pairplot. Based on this plot what looks to be the most correlated feature with 
the Yearly Amount Spent?
"""
sns.set_style('ticks')
sns.pairplot(df)

# Length_of_Membership is the most correlated one
```

```python
"""
Compute - Create a linear model plot of the Length of Membership and Yearly 
Amount Spent. Does the data fit well in a linear plot?
"""

x = df['Length_of_Membership']
y = df['Yearly_Amount_Spent']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=85)

lrm = LinearRegression()
lrm.fit(pd.DataFrame(x_train), y_train)

"""
Compute - Predict the data and do a scatter plot. Check if actual and predicted 
data match.
"""

predicted = lrm.predict(pd.DataFrame(x_test))
plt.scatter(predicted, y_test)
plt.show()
```

```python
# What is the value of Root Mean Squared Error?
print(mean_squared_error(y_test, predicted, squared=False))
```
