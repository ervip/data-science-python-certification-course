```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\CourseWork\data-science-python-certification-course\Assignments\04 Introduction to NumPy, Pandas, and Matplotlib\Case Study IV\resources\BigMartSalesData.csv')
```

```python
# Answer 1

tdf = df[df['Year'] == 2011].filter(['Month', 'Amount']).groupby('Month', as_index=False).sum()

plt.plot(tdf['Month'], tdf['Amount'])
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Answer 2

plt.bar(tdf['Month'], tdf['Amount'])
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Answer 3

tdf = df[df['Year'] == 2011].filter(['Country', 'Amount']).groupby('Country', as_index=False).sum()

plt.pie(tdf['Amount'], labels=tdf['Country'], shadow=True, autopct='%1.1f%%')
plt.show()
```

```python
# Answer 4

tdf = df.filter(['InvoiceDate', 'Amount']).groupby('InvoiceDate', as_index=False).sum()

plt.scatter(tdf['InvoiceDate'], tdf['Amount'])
plt.show()
```
