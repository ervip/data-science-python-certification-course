```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Phase 1

CSV_PATH = r"D:\CourseWork\data-science-python-certification-course\Assignments\04 Introduction to NumPy, Pandas, and Matplotlib\Case Study I\resources\middle_tn_schools.csv"
df = pd.read_csv(CSV_PATH)
print(df.describe())
```

```python
# Phase 2

df.drop('reduced_lunch', axis=1).groupby(by='school_rating').describe()
```

```python
# Phase 3

df['reduced_lunch'].corr(df['school_rating'])
```

```python
# Phase 4

plt.scatter(x=df['reduced_lunch'], y=df['school_rating'])
plt.xlabel('reduced_lunch =>')
plt.ylabel('school_rating =>')
plt.show()
```

```python
# Phase 5

plt.matshow(df.drop(['name', 'school_type'], axis=1).corr())
plt.show()
```
